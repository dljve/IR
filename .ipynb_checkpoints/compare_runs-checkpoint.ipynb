```python
def load_metrics(file):
    metrics = {}
    with open(file, 'r') as f:
        for line in f:
            metric, qid, score = line.split('\t')
            metric = metric.strip()
            qid = qid.strip()
            score = score.strip()
            if qid == 'all':
                continue
            if metric not in metrics:
                metrics[metric] = {}
            metrics[metric][qid] = float(score)

    return metrics
```

```python
def plot(all_results, ymin=-1, ymax=1, output_path="."):
    fig, ax = plt.subplots(1, 1, figsize=(16, 3))
    all_results.sort(key = itemgetter(1), reverse=True)
    x = [_x+0.5 for _x in range(len(all_results))]
    y = [float(ele[1]) for ele in all_results]
    ax.bar(x, y, width=0.6, align='edge')
    ax.set_xticks(x)
    ax.set_xticklabels([int(ele[0]) for ele in all_results], {'fontsize': 4}, rotation='vertical')
    ax.grid(True)
    ax.set_title("Per-topic analysis on {}".format(metric))
    ax.set_xlabel('Topics')
    ax.set_ylabel('{} Diff'.format(metric))
    ax.set_ylim(ymin, ymax)
    plt.show()
    #output_fn = os.path.join(output_path, 'per_query_{}.pdf'.format(metric))
    #plt.savefig(output_fn, bbox_inches='tight', format='pdf')
```

```python
def compare_values(b,c,m):
    base = "C:/Users/Loes/Documents/M1/IR/IR/runs2/" + b
    comp = "C:/Users/Loes/Documents/M1/IR/IR/runs2/" + c
    qrels = "C:/Users/Loes/Documents/M1/IR/anserini/src/main/resources/topics-and-qrels/qrels.msmarco-doc.dev.txt"
    metric = m
    msmarco = False
    ymin = -1
    ymax = 1
    
    if msmarco:
        base_all, base_metrics = compute_metrics_from_files(qrels, base, per_query_score=True) 
        comp_all, comp_metrics = compute_metrics_from_files(qrels, comp, per_query_score=True) 
    else:
        os.system(f'C:/Users/Loes/Documents/M1/IR/IR/trec_eval -q -M1000 -m {metric} {qrels} {base} > eval.base')
        os.system(f'C:/Users/Loes/Documents/M1/IR/IR/trec_eval -q -M1000 -m {metric} {qrels} {comp} > eval.comp')
        
    base_metrics = load_metrics('eval.base')
    comp_metrics = load_metrics('eval.comp')
    
    if "." in metric:
        metric = "_".join(metric.split("."))

    if msmarco:
        metric = "MRR@10"
    keys = []
    for key in base_metrics[metric]:
        base_score = base_metrics[metric][key]
        if key not in comp_metrics[metric]:
            continue
        keys.append(key)

    # Extract the paired scores
    a = [base_metrics[metric][k] for k in keys]
    b = [comp_metrics[metric][k] for k in keys]

    (tstat, pvalue) = scipy.stats.ttest_rel(a, b)
    w, p = wilcoxon(a,b)
    base_mean = np.mean(a)
    comp_mean = np.mean(b)
    t_pval = pvalue
    w_pval = p
    return comp_mean, t_pval, w_pval

numbers = [1,2,3]
words = ["synonyms", "hypernyms", "hyponyms", "meronyms", "holonyms"]
w = "synonyms"

def make_plott(n,metric,weight):   
    comp_means = []
    t_pvals = []
    w_pvals = []
    for w in words:
        x = "wordnet_"+w+str(n)+weight+".txt"
        comp_mean, t_pval, w_pval = compare_values("bm25.txt",x,metric)
        comp_means.append(comp_mean)
        t_pvals.append(t_pval)
        w_pvals.append(w_pval)
    return comp_means, t_pvals, w_pvals

def metric_plot(metric,bot,top,weight):
    comp_means1, t_pvals1, w_pvals1 = make_plott(1,metric,weight)
    comp_means2, t_pvals2, w_pvals2 = make_plott(2,metric,weight)
    comp_means3, t_pvals3, w_pvals3 = make_plott(3,metric,weight)

    labels = ["synonyms", "hypernyms", "hyponyms", "meronyms", "holonyms"]
    
    highest = []
    words2 = []
    for w in range(len(labels)):
        temp = [comp_means1[w],comp_means2[w],comp_means3[w]]
        highi = np.argmax(temp)
        st = labels[w] + str(highi+1)
        words2.append(st)
        highest.append(temp[highi])
    
    outcome = metric + ": " + str(words2) + weight

    x = np.arange(len(labels))  # the label locations
    width = 0.2  # the width of the bars

    fig, ax = plt.subplots()
    rects1 = ax.bar(x - width, comp_means1, width, label='Comp1')
    rects2 = ax.bar(x, comp_means2, width, label='Comp2')
    rects3 = ax.bar(x + width, comp_means3, width, label='Comp3')
    m = ""
    if metric is "map":
        m = "MAP"
    elif metric is "ndcg":
        m = "NDCG"
    else:
        m = "MRR"
    
    weight_val = ""
    if weight is "_01":
        weight_val = "0.1"
    elif weight is "_03":
        weight_val = "0.3"
    else:
        weight_val = "0.5"
    
    ax.set_ylabel(m)
    title = "Varying number of additions over words for " + m + " with weight " + weight_val
    ax.set_title(title)
    ax.set_xticks(x)
    ax.set_xticklabels(labels)
    ax.legend(loc=4)
    ax.set_ylim(bot,top)
    
metric_plot("map",0.13,0.28,"_01")
metric_plot("ndcg",0.13,0.28,"_01")
metric_plot("recip_rank",0.13,0.28,"_01")

metric_plot("map",0.13,0.28,"_03")
metric_plot("ndcg",0.13,0.28,"_03")
metric_plot("recip_rank",0.13,0.28,"_03")

metric_plot("map",0.13,0.28,"_05")
metric_plot("ndcg",0.13,0.28,"_05")
metric_plot("recip_rank",0.13,0.28,"_05")
```

```python
#highest singles
def comps(metric,bot,top):
    bscore = []
    bt_vals = []
    bw_vals = []
    rscore = []
    rt_vals = []
    rw_vals = []
    
    tests = ["bm25","bm25+rm3"]
    #tests.extend(["wordnet_synonyms1","wordnet_hypernyms2","wordnet_hyponyms1","wordnet_meronyms2","wordnet_holonyms1"])
    tests.extend(["wordnet_hyper+hypo","wordnet_mero+holo","wordnet_syn+hyper+hypo","wordnet_syn+mero+holo","wordnet_all"])
    
    for x in tests:
        value, t_pval, w_pval = compare_values("bm25.txt",x+".txt",metric)
        bscore.append(value)
        bt_vals.append(t_pval)
        bw_vals.append(w_pval)
        value, t_pval, w_pval = compare_values("bm25+rm3.txt",x+".txt",metric)
        rscore.append(value)
        rt_vals.append(t_pval)
        rw_vals.append(w_pval)
    
    labels = ["bm25","bm25+rm3","hyper+hypo","mero+holo","syn+hyper+hypo","syn+mero+holo","all"]
    
    print("metric", metric)
    print("bscore", bscore)
    print("bt_vals", bt_vals)
    print("bw_vals", bw_vals)
    print("rscore", rscore)
    print("rt_vals", rt_vals)
    print("rw_vals", rw_vals)
    
    m = ""
    if metric is "map":
        m = "MAP"
    elif metric is "ndcg":
        m = "NDCG"
    else:
        m = "MRR"
    
    plt.figure(figsize=(15,5))
    plt.bar(labels,bscore,color="tab:blue")
    plt.ylabel(m)
    title = "The " + m + " score for bm25, bm25+rm3 and combinations of wordnet expansions"
    plt.title(title)
    plt.ylim(bot,top)
    plt.show()
    
comps("map",0.14,0.28)
comps("ndcg",0.14,0.28)
comps("recip_rank",0.14,0.28)
```
